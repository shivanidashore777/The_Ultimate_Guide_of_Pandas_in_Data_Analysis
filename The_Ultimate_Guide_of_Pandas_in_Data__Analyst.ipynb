
**Advantages and differences of a Pandas Series compared to a 1-D NumPy array:**

**Generalization and Flexibility:** The Pandas Series is more general and flexible compared to a 1-D NumPy array. It provides additional capabilities and functionality beyond what a NumPy array offers.

 **Series as a Generalized NumPy Array:** The Pandas Series emulates a 1-D NumPy array, which means it can perform many similar operations and computations. It can be used as a substitute for a NumPy array in many cases.

**Presence of Index:** The significant difference between a Series and a 1-D NumPy array is the presence of an index in the Series. The index is associated with the values and provides a clear-cut definition of the relationship between the values and their corresponding labels.

**Explicitly Defined Index:** Unlike a NumPy array that has an implicitly defined integer index, the Pandas Series has a well-defined and explicit index associated with the values. This index allows for easy and efficient value retrieval based on the index labels.

 **Non-Integer Index:**The index in a Pandas Series can be composed of values of any desired type. It is not limited to integers only. For example, you can use strings as an index, allowing for more meaningful and descriptive labels associated with the values.

In summary, the Pandas Series is a more versatile data structure than a 1-D NumPy array because of its explicit and flexible index, which provides additional capabilities and allows for more meaningful labeling of values.

```
exam1
```

Pandas provides various methods for cleaning the missing values. The fillna function can “fill in” NA values with non-null data in a couple of ways, which we have illustrated in the following sections.

```
exam2=exam.fillna(0) #replace NaN values with 0
exam2
```

```
# fill the Nan values in math column with median of the column

a=exam['Math'].median()
a
exam1['Math']=exam['Math'].fillna(a)
exam1
```

```
# fill the Nan values in phy column with mean of the column
b=exam1['Phy'].mean()
b
exam1['Phy']=exam1['Phy'].fillna(b)
exam1
```

# Creating joints in Pandas
Full Outer Join

combines the results of both the left and the right outer joins. The joined DataFrame will contain all records from both the DataFrames and fill in NaNs for missing matches on either side. You can perform a full outer join by specifying the how argument as outer in the merge() function:

Inner Join

combines the common results of both

There should be relevance(Common field,analysis requirement)

```
import pandas as pd
d = {
        'id': ['1', '2', '3', '4', '5','6'],
        'Color': ['RED', 'GREEN', 'YELLOW', 'BLUE', 'PINK','BLACK'],
        'Fruit': ['APPLE', 'BANANA', 'MANGO', 'BERRY', 'MELON','GRAPES']}
d1=pd.DataFrame(d)
d1
```

```
z = {    'id': ['1', '2', '3', '4', '5','7'],
        'rating': ['A', 'B', 'C', 'A', 'B','A'],
        'COST': [200, 230, 400, 400, 100,450],
        'Fruit': ['APPLE', 'BANANA', 'MANGO', 'BERRY', 'MELON','KIWI'],
        'BUY': ['Y', 'N', 'Y', 'N', 'Y','N']}
d2=pd.DataFrame(z)
d2
```

```
join=pd.merge(d1,d2,on='Fruit',how='right') # similary we can  use 'outer/right/left join'
join
```

```
#Selecting few fields from frames
join=pd.merge(d1[["Color","Fruit"]],d2,on='Fruit',how='outer') ## similary we can  use 'inner/right/left join'
join
```

```
#Selecting both common fields from frames
join = pd.merge(d1, d2, on=["id","Fruit"], how='outer') ## similary we can  use 'inner/right/left join'

join
```
